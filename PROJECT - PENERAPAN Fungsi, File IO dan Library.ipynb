# Total Nilai Impor Tahun 2005 - 2013 (dalam ribuan USD)

```python
total_impor = data_impor.nilai_impor.sum()
print("Total Nilai Impor Tahun 2005 - 2013 (dalam ribuan US$) : ", total_impor)
```

# Nilai Impor Berdasarkan Region Tahun 2005 - 2013 (dalam ribuan USD)

```python
data_impor.groupby('region').sum()['nilai_impor']
```

# Tampilan data nilai impor berdasarkan region

```python
ax = data_impor.groupby('region').sum()['nilai_impor']
ax = ax.plot(kind='bar', title ="Nilai Impor Berdasar Negara", legend=True, fontsize=12)
```

# Analisis data nilai impor berdasarkan region

Berdasarkan grafik bar diatas, pada tahun 2005 hingga 2013, impor tertinggi indonesia adalah ke region asia timur, diikuti dengan region ASEAN dan Uni Eropa serta Amerika Utara. Selain region tersebut, data impor indonesia signifikan jauh dibawah region-region yang telah disebutkan sebelumnya, sehingga dapat dikatakan bahwa indonesia, sangat bergantung pada wilayah Asia Timur untuk masalah impor.

#  Tampilan data nilai impor berdasarkan tahun

```python
ax = data_impor.groupby('tahun').sum()['nilai_impor']
ax = ax.plot(kind='line', title ="Nilai Impor Berdasar Negara", legend=True, fontsize=12)
```

# Analisis data nilai impor berdasarkan tahun

Berdasarkan grafik line diatas, dapat disimpulkan bahwa pada tahun 2005 hingga 2008 nilai impor indonesia mengalami kenaikan. Namun, pada tahun 2008- 2009, nilai impor Indonesia mengalami penurunan yang sangat signifikan. Setelah itu, pada periode 2009- 2012,  nilai impor yang dilakukan yang dilakukan Indonesia terus mengalami kenaikan dan mengalami sedikit penurunan pada tahun 2013.
